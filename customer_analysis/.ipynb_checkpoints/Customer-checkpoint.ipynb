```python
# keep the last two columns
country_cust_data = Rt1_data[['Country','CustomerID']].drop_duplicates()
country_cust_data.groupby(['Country'])['CustomerID'].aggregate('count').reset_index().sort_values('CustomerID', ascending=False)
```

```python
# filter data of UK
Rt1_data = Rt1_data.query("Country=='United Kingdom'").reset_index(drop=True)
```

```python
# check for missing values
Rt1_data.isnull().sum(axis = 0)
```

```python
# remove missing values from CustomerID column, can ignore values in description column
Rt1_data = Rt1_data[pd.notnull(Rt1_data['CustomerID'])]

# validate if there are any negative values in Quantity column
Rt1_data.Quantity.min()
```

```python
# validate if there are any negative values in UnitPrice column
Rt1_data.UnitPrice.min()
```

```python
# filter out records with negatives values
Rt1_data = Rt1_data[(Rt1_data['Quantity']>0)]
```

```python
# convert type string to datetime
Rt1_data['InvoiceDate'] = pd.to_datetime(Rt1_data['InvoiceDate'])
```

```python
# add new column for totalamount
Rt1_data['TotalAmount'] = Rt1_data['Quantity']*Rt1_data['UnitPrice']
```

```python
import datetime as dt

# set latest dat 2011-12-10 as last invoice date was 2011-12-9. This is to calculate the number of days from recent purchase
Latest_Date = dt.datetime(2011,12,10)

# create RFM Modeling scores for each customer
RFMScores = Rt1_data.groupby('CustomerID').agg({'InvoiceDate': lambda x: (Latest_Date - x.max()).days, 'InvoiceNo': lambda x: len(x), 'TotalAmount': lambda x: x.sum()})

# convert Invoice date into type int
RFMScores['InvoiceDate'] = RFMScores['InvoiceDate'].astype(int)

# rename column to Recency, Frequency, Monetary
RFMScores.rename(columns={'InvoiceDate':'Recency', 'InvoiceNo':'Frequency', 'TotalAmount':'Monetary'}, inplace = True)

RFMScores.reset_index().head()
```

```python
# split into four segments using quantiles
quantiles = RFMScores.quantile(q=[.25, .5, .75])
quantiles = quantiles.to_dict()
```

```python
quantiles
```

```python
# for R, the smaller the better
def RScoring(x, p, d):
    if x <= d[p][.25]:
        return 1
    elif x <= d[p][.5]:
        return 2
    elif x <= d[p][.75]:
        return 3
    else:
        return 4

# for F and M, the larger the better
def FnMScoring(x, p, d):
    if x <= d[p][.25]:
        return 4
    elif x <= d[p][.5]:
        return 3
    elif x <= d[p][.75]:
        return 2
    else:
        return 1
 
```

```python
# add a new column of R
RFMScores['R'] = RFMScores['Recency'].apply(RScoring, args=('Recency', quantiles,))
# add a new column of F
RFMScores['F'] = RFMScores['Frequency'].apply(FnMScoring, args=('Frequency', quantiles,))
# add a new column of M
RFMScores['M'] = RFMScores['Monetary'].apply(FnMScoring, args=('Monetary', quantiles,))
RFMScores.head()
```

```python
# add a new column for the series of RFM
RFMScores['RFMGroup'] = RFMScores.R.map(str) + RFMScores.F.map(str) + RFMScores.M.map(str)
# add a new column to sum up RFM
RFMScores['RFMScore'] = RFMScores[['R','F','M']].sum(axis = 1)
RFMScores.head()
```

```python
# assign Loyalty level to each customer
Loyalty_Level = ['Platinm','Gold','Silver','Bronze']
Score_cuts = pd.qcut(RFMScores.RFMScore, q = 4, labels = Loyalty_Level)
RFMScores['RFM_Loyalty_Level'] = Score_cuts.values
RFMScores.reset_index().head()
```

```python
# validate the data for RFMGroup = 111
RFMScores[RFMScores['RFMGroup']=='111'].sort_values('Monetary', ascending=False).reset_index().head(10)
```

```python
# assign zero and negative values to one
def handle_neg_n_zero(num):
    if num <= 0:
        return 1
    else:
        return num
# apply above function to Recency and Monetary columns
RFMScores['Recency'] = [handle_neg_n_zero(x) for x in RFMScores.Recency]
RFMScores['Monetary'] = [handle_neg_n_zero(x) for x in RFMScores.Monetary]
# perform log transformation to bring data into normal or neor normal distribution
Log_Tfd_Data = RFMScores[['Recency','Frequency','Monetary']].apply(np.log, axis = 1).round(3)
```

```python
# plot the processed Recency data
Recency_Plot = Log_Tfd_Data['Recency']
ax = sns.distplot(Recency_Plot)
```

```python
# plot the processed Frequency data
Frequency_Plot = Log_Tfd_Data['Frequency']
ax = sns.distplot(Frequency_Plot)
```

```python
# plot the processed Monetary data
Monetary_Plot = Log_Tfd_Data['Monetary']
ax = sns.distplot(Monetary_Plot)
```

```python
from sklearn.preprocessing import StandardScaler

# bring the data on the same scale
scaleobj = StandardScaler()
Scaled_Data = scaleobj.fit_transform(Log_Tfd_Data)

# transform it back to dataframe
Scaled_Data = pd.DataFrame(Scaled_Data, index = RFMScores.index, columns = Log_Tfd_Data.columns)
```

```python
from sklearn.cluster import KMeans

sum_of_sq_dist = {}
for k in range(1,15):
    km = KMeans(n_clusters = k, init = 'k-means++', max_iter = 1000)
    km = km.fit(Scaled_Data)
    sum_of_sq_dist[k] = km.inertia_
    
# plot the graph for the sum of square distance values and number of clusters
sns.pointplot(x=list(sum_of_sq_dist.keys()), y=list(sum_of_sq_dist.values()))
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Sum of Square Distances')
plt.title('EIbow Method For Optimal k')
plt.show()
```

```python
# perform K-Mean clustering model when k=3
KMean_clust = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 1000)
KMean_clust.fit(Scaled_Data)

# add a new column to RFMScores
RFMScores['Cluster'] = KMean_clust.labels_
RFMScores.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7,7))

# scatter plot Frequency vs Recency
colors = ['red','green','blue']
RFMScores['Color'] = RFMScores['Cluster'].map(lambda p: colors[p])
ax = RFMScores.plot(
    kind='scatter',
    x='Recency',
    y='Frequency',
    figsize=(10,8),
    c= RFMScores['Color']
)
```
