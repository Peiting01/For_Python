```python
import matplotlib.pyplot as plt
import pandas_datareader.data as pdr
import datetime as dt
import numpy as np
import yfinance as yf
yf.pdr_override()
```

```python
tickers = ['AAPL', 'HPE', 'FB', 'v']
weights = np.array([.20, .20, .20, .20])

# download dataframe
data = pdr.get_data_yahoo(tickers, start="2019-01-01", end=dt.datetime.today())['Close']

# calculate returns
returns = data.pct_change()
print(returns)
# create covariance matrix
cov_matrix = returns.cov()
print(cov_matrix)
# calculate mean for each stock
avg_rets = returns.mean()
print(avg_rets)
# calculate portfolio mean
port_mean = avg_rets.dot(weights)
print(port_mean)
# calculate portfolio stdev
port_stdev = np.sqrt(weights.T.dot(cov_matrix).dot(weights))
print(port_stdev)
# define investment
investment = 100000

# port mean($)
mean_investment = (1 + port_mean) * investment
print(mean_investment)
# port stdev($)
stdev_investment = investment * port_stdev
print(stdev_investment)
# confidence level
conf_level = 0.01

#cutoff
from scipy.stats import norm
cut = norm.ppf(conf_level, mean_investment, stdev_investment)
print(cut)
# calculate var
var_id = investment - cut
print(var_id)
#calculate var over different frequencies
num_days = 51

for x in range(1, num_days):
    print (str(x) + ' Day_Var: ' + str(np.round(var_id * np.sqrt(x), 2)))

```
