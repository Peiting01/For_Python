```python
import pandas_datareader.data as web
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from matplotlib import style
from mpl_finance import candlestick_ohlc
```

```python
style.use('ggplot')

start1 = dt.datetime(2015, 1, 3)
end1 = dt.datetime(2020, 3, 31)

google_df = web.DataReader('GOOGL','yahoo', start1, end1) 
print(google_df.head(3))

# google_df['50ma'] = google_df['Adj Close'].rolling(window=50, min_periods=0).mean()
google_df_ohlc = google_df['Adj Close'].resample('10D').ohlc()
google_df_volume = google_df['Volume'].resample('10D').sum()
print(google_df_ohlc.head())
# set date
google_df_ohlc.reset_index(inplace=True)
google_df_ohlc['Date'] = google_df_ohlc['Date'].map(mdates.date2num)
print(google_df_ohlc.head())
# set the picture grids
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)
# plot on the grids
ax1.plot(google_df.index, google_df['Adj Close'])
ax2.bar(google_df.index, google_df['Volume'], color='blue')

plt.show()
```

```python
# set the picture grids
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)
ax1.xaxis_date()
candlestick_ohlc(ax1, google_df_ohlc.values, width=2, colorup='g')
ax2.fill_between(google_df_volume.index.map(mdates.date2num), google_df_volume.values, 0, color='blue')

plt.show()
```
